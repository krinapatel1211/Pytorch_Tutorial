## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - Returns a view of an input tensor of certain shape but keep the same memory as the original tensor. changing the new tensor will also change the original tensor
* stacking - combine multiple tensors on top of one another (vstck) or side by side (hstack)
* Squeeze = removes all '1' dimension to a target tensor
* Unsqueeze - add dimension to existing tensor to a specific dimension
* permutes - swap order or dimensions

```python
x = torch.arange(1., 10.)
x.shape, x
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
z = x.view(1,9)
z, z.shape
```

```python
z[:,0] = 5
z,x
```

```python
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked
```

```python
x_stacked = torch.stack([x,x,x,x], dim=1)
x_stacked
```

```python
x_hstacked = torch.hstack([x,x,x,x,x,x,x])
x_hstacked
```

```python
x_vstacked = torch.vstack([x,x,x,x,x,x,x,x])
x_vstacked
```

```python
x = torch.zeros([1,2,1,4])
x
```

```python
y = x.squeeze()
y.shape
```

```python
z = y.unsqueeze(dim=1)
z.shape
```

```python
w = z.unsqueeze(dim=3)
w.shape
```

## PyTorch tensors and Numpy
* Data in NumPy, want in PyTorch tensor --> `torch.from_numpy(ndarray`
* PyTorch tensor --> NumPy --> `torch.Tensor.numpy()`

```python
array = np.arange(1.0,8.0)
array.dtype
```

```python

tensor = torch.from_numpy(array) # when converting from numpy to pytorch, pytorch
array, tensor
```

```python
array = array+1
array, tensor
```

```python
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
tensor = tensor +1
tensor, numpy_tensor
```
